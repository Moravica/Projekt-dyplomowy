```python
from flask import Flask, render_template, request, redirect, url_for
import threading
import webbrowser
import threading

webbrowser.open("http://127.0.0.1:5000/howdy")

app = Flask(__name__)

opcje = [f"Kadencja {i}" for i in range(3, 11)]
opcje2 = [f"Kadencja {i}" for i in range(7, 11)]
opcje3 = [f"Kadencja {i}" for i in range(8, 11)]
opcje4 = [f"Kadencja {i}" for i in range(1, 10)]+["Wszystkie kadencje"]
opcje5 = ["Rok urodzenia","Miejsce urodzenia","Poziom wykształcenia","Ukończona szkoła","Zawód"]
start_time = None

def execute_notebook(notebook_number):
    if notebook_number == 1:
        %run ustawy.ipynb
    elif notebook_number == 2:
        %run terminy.ipynb
    elif notebook_number == 3:
        %run wyszukiwarka.ipynb
    elif notebook_number == 4:
        %run poslowie.ipynb

    webbrowser.open("http://127.0.0.1:5000/howdy")
    
@app.route("/howdy")
def howdy():
    return render_template('index.html', opcje=opcje)

@app.route("/poslowie", methods=['GET', 'POST'])
def poslowie():
    wybrana_opcja=""
    if request.method == 'POST':
        wybrana_opcja = request.form['opcja']
        
        print("Wybrano:", wybrana_opcja)
        
        with open('przekazana_zmienna.txt', 'w') as f:
            f.write(wybrana_opcja)

        # Uruchamianie funkcji w oddzielnym wątku
        notebook_thread = threading.Thread(target=execute_notebook, args=(4,))
        notebook_thread.start()

        return render_template('loader.html', opcje=opcje4, wybrana_opcja=wybrana_opcja)
        # Tutaj możesz wykonać dowolne działania na wybranej opcji
    return render_template('poslowie.html', opcje=opcje4, wybrana_opcja=wybrana_opcja) 

@app.route("/wyszukaj-wypowiedz", methods=['GET', 'POST'])
def wyszukaj():
    wybrana_opcja=""
    if request.method == 'POST':
        wybrana_opcja = request.form['opcja']
        wpisany_tekst = request.form['tekst']
        with open('przekazana_zmienna.txt', 'w') as f:
            f.write(wybrana_opcja)
        with open('przekazany_tekst.txt', 'w') as f:
            f.write(wpisany_tekst)

        # Uruchamianie funkcji w oddzielnym wątku
        notebook_thread = threading.Thread(target=execute_notebook, args=(3,))
        notebook_thread.start()
        print(wybrana_opcja)

        return render_template('loader.html', opcje=opcje3, wybrana_opcja=wybrana_opcja)
        # Tutaj możesz wykonać dowolne działania na wybranej opcji
    

    return render_template('wyszukaj-wypowiedz.html', opcje=opcje3, wybrana_opcja=wybrana_opcja) 

@app.route("/wyszukaj_posla", methods=['GET', 'POST'])
def wyszukaj_posla():
    wybrana_opcja=""
    wybrana_opcja2=""
    if request.method == 'POST':
        wybrana_opcja = request.form['opcja']
        wybrana_opcja2 = request.form['opcja2']
        wpisany_tekst = request.form['tekst']

        print(wybrana_opcja)
        print(wybrana_opcja2)
        print(wpisany_tekst)
        print("Wybrano:", wybrana_opcja, wybrana_opcja2)
        with open('szukana_wartosc.txt', 'w') as f:
            f.write(wybrana_opcja2+";"+wpisany_tekst+";"+wybrana_opcja[-1])


        # Uruchamianie funkcji w oddzielnym wątku
        notebook_thread = threading.Thread(target=execute_notebook, args=(4,))
        notebook_thread.start()
        print(wybrana_opcja)

        return render_template('loader.html', opcje=opcje4,opcje2=opcje5, wybrana_opcja=wybrana_opcja)
        # Tutaj możesz wykonać dowolne działania na wybranej opcji
    

    return render_template('wyszukaj_posla.html', opcje=opcje4,opcje2=opcje5, wybrana_opcja=wybrana_opcja) 

@app.route("/komisje", methods=['GET', 'POST'])
def komisje():
    wybrana_opcja=""
    if request.method == 'POST':
        wybrana_opcja = request.form['opcja']
        
        print("Wybrano:", wybrana_opcja)
        
        with open('przekazana_zmienna.txt', 'w') as f:
            f.write(wybrana_opcja)

        # Uruchamianie funkcji w oddzielnym wątku
        notebook_thread = threading.Thread(target=execute_notebook, args=(2,))
        notebook_thread.start()

        return render_template('loader.html', opcje=opcje2, wybrana_opcja=wybrana_opcja)
        # Tutaj możesz wykonać dowolne działania na wybranej opcji
    return render_template('komisje.html', opcje=opcje2, wybrana_opcja=wybrana_opcja) 

@app.route("/ustawy", methods=['GET', 'POST'])
def ustawy():
    wybrana_opcja=""
    if request.method == 'POST':
        wybrana_opcja = request.form['opcja']
        print("Wybrano:", wybrana_opcja)
        with open('przekazana_zmienna.txt', 'w') as f:
            f.write(wybrana_opcja)

        # Uruchamianie funkcji w oddzielnym wątku
        notebook_thread = threading.Thread(target=execute_notebook, args=(1,))
        notebook_thread.start()

        return render_template('loader.html', opcje=opcje, wybrana_opcja=wybrana_opcja)
        # Tutaj możesz wykonać dowolne działania na wybranej opcji
    return render_template('ustawy.html', opcje=opcje, wybrana_opcja=wybrana_opcja) 

if __name__ == "__main__":
    app.run(port=5000)
```
